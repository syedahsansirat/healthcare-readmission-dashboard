# Healthcare Performance & Readmission Analytics Dashboard

This project aims to analyze hospital readmission patterns and key performance indicators (KPIs) to support data-driven decision-making in healthcare settings. Using patient-level data, the objective is to identify trends in readmissions, evaluate factors such as patient demographics, length of stay, and admission characteristics, and highlight potential areas for improving care quality and operational efficiency. The project involves data cleaning and transformation using Python (Pandas), creation of meaningful KPIs such as readmission rates and patient distribution, and preparation of a structured dataset for visualization in Power BI. The final output focuses on delivering clear insights through dashboards to support performance monitoring and resource planning, aligned with healthcare reporting practices such as those used by CIHI.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = '/content/hospital_readmissions.csv'
df = pd.read_csv(file_path)

print("Dataset loaded successfully. Displaying the first 5 rows:")
print(df.head())

print("\nBasic information about the dataset:")
df.info()

print("\nDescriptive statistics for numerical columns:")
print(df.describe())

print("\nMissing values in each column:")
print(df.isnull().sum())
```

```python
# Total patients
total_patients = len(df)

# Readmission rate
readmission_rate = df['readmitted'].value_counts(normalize=True) * 100

# Calculate Length of Stay from 'time_in_hospital' column
df['length_of_stay'] = df['time_in_hospital']
avg_los = df['length_of_stay'].mean()
```

Creating Some features

```python
# Extract the numerical lower bound from the 'age' string column
df['age_numeric'] = df['age'].str.extract(r'\[(\d+)-').astype(int)

df['age_group'] = pd.cut(df['age_numeric'], bins=[0,30,50,70,100], labels=['Young','Adult','Senior','Elder'])
```

Export Clean Data

```python
df.to_csv("cleaned_readmission_data.csv", index=False)
```
